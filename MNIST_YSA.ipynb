# Yapay Sinir Ağı Mimarisi

```python
from keras import models
from keras import layers

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
network = models.Sequential() #Boş bir model oluşturulması
network.add(layers.Dense(512, activation = 'relu', input_shape = (28 * 28,))) #Girişler 28*28 boyutunda olduğu için input_shape 28*28 olur
network.add(layers.Dense(10, activation = 'softmax')) #10 adet rakam olduğu için 10 sınıf olacak
```

## YSA Modelinin Derlenmesi

```python
network.compile(optimizer = 'rmsprop',
                loss = 'categorical_crossentropy',
                metrics = ['accuracy'])
```

##Girişlerin ve Etiketlerin Hazırlanması

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255 #Görüntü işlemede her zaman uygulanan bir normalizasyon işlemi

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255 #Görüntü işlemede her zaman uygulanan bir normalizasyon işlemi
```

```python
from keras.utils import to_categorical

train_labels = to_categorical(train_labels) #Categorical hale getirilmiş train_label'lar tekrar train_label'lara atanarak üzerine yazılır
test_labels = to_categorical(test_labels)
```

## YSA Modelinin Eğitilmesi

```python
network.fit(train_images, train_labels, epochs=10, batch_size = 128)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
```

```python
print('test_loss: ', test_loss)
print('test_acc: ', test_acc) #Doğruluk oranı
```
